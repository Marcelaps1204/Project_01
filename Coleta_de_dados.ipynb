 ### 1. Selecionar os números do cartão de fidelidade dos passeiros, a cidade e o gênero, dos passageiros que tem o cartão Star de fidelidade, mas nunca realizaram nenhuma viagem de avião ( Dica: “loyalty_card” = “Star” e “distance” = 0 )

```python
cosulta_atividade = """
  SELECT
    fa.loyalty_number,
    flh.city,
    flh.gender,
    flh.loyalty_card
  FROM flight_activity fa LEFT JOIN flight_loyalty_history flh
                              ON (fa.loyalty_number = flh.loyalty_number)
  WHERE flh.loyalty_card = 'Star' AND fa.distance = '0'
"""
df_atividade = pd.read_sql_query (consulta_atividade, conn)
df_atividade.head(10)
```

### 2) Selecionar os números do cartão de fidelidade, o gênero e a cidade de todos os passageiros do sexo feminino que moram na cidade de Toronto, fizeram mais de 30 viagens no total e tem o cartão de fidelidade do tipo Aurora.

```python
cosulta_atividade= """
  SELECT
    fa.loyalty_number,
    flh.city,
    flh.gender,
    flh.loyalty_card
  FROM flight_activity fa LEFT JOIN flight_loyalty_histoy flh
                            ON (fa.loyalty_number = flh.loyalty_number)
  WHERE flh.loyalty_card = 'Aurora'
            AND fa.gender = 'Female'
            AND flh.city = 'Toronto'
            AND flh.total_flights > '30'
"""
df_atividade = pd.read_sql_query (consulta_atividade, conn)
df_atividade.head(10)
```
